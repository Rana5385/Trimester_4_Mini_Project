```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../data/test_image.jpg')
lane_image = np.copy(image)
gray = cv2.cvtColor(lane_image, cv2.COLOR_BGR2GRAY)
```

```python
print(f'Before:\t{image.shape=}\t{gray.shape=}\t{lane_image.shape=}')
image = cv2.resize(image, (int(image.shape[1] * .5), int(image.shape[0] * .5)))
gray = cv2.resize(gray, (int(gray.shape[1] * .5), int(gray.shape[0] * .5)))
print(f'After:\t{image.shape=}\t{gray.shape=}\t{lane_image.shape=}')
```

```python
plt.imshow(image)
```

```python
plt.imshow(gray)
```
